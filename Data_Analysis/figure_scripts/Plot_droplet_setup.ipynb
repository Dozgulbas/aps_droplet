## Initialize

```python
import cv2
from matplotlib import pyplot as plt
from mpl_toolkits.axes_grid1.inset_locator import inset_axes
```

## Reading images

```python
# reading images
Image1 = cv2.imread('../Data/IMG_6882.png')
Image2 = cv2.imread('../Data/drop_close.jpg')

Image1 = cv2.cvtColor(Image1, cv2.COLOR_BGR2RGB)
Image2 = cv2.cvtColor(Image2, cv2.COLOR_BGR2RGB)
```

## Setting figure size

```python
scale = 1
width = 3.6*scale
font_size = 8*scale
line_width = 0.5*scale
marker_size = 1*scale
tick_length_major = 4*scale
tick_length_minor = 2*scale
dim=1
```

## Plot droplet setup figure

```python
# create figure
fig, ax = plt.subplots(1,1,figsize=(width, 0.8*width))  
# setting values to rows and column variables
rows = 1
columns = 2
ax.imshow(Image1)

# Setting the ticks
ax.tick_params('both', length=tick_length_major, width=line_width, which='major', labelsize=font_size)
ax.tick_params('both', length=tick_length_minor, width=0.5*line_width, which='minor',
                labelleft=False, labelbottom=False)
ax.tick_params(which='major' ,top=False, bottom=False, left=False, right=False,
                    labelleft=False, labelbottom=False)

# Creating an inset image
axins = inset_axes(ax, width="100%", height="100%",bbox_to_anchor=(0.578,1-0.37,0.42,0.42), bbox_transform=ax.transAxes, borderpad=0.1)
axins.imshow(Image2)
for spine in axins.spines.values():
        spine.set_edgecolor('yellow')

axins.tick_params('both', length=tick_length_major, width=line_width, which='major', labelsize=font_size)
axins.tick_params('both', length=tick_length_minor, width=0.5*line_width, which='minor',
                labelleft=False, labelbottom=False)
axins.tick_params(which='major' ,top=False, bottom=False, left=False, right=False,
                    labelleft=False, labelbottom=False)

plt.savefig('../figures/exp_setup.pdf', dpi=600, format='pdf', 
        facecolor='w', edgecolor='w', transparent=False, bbox_inches='tight', pad_inches = 0)

```
